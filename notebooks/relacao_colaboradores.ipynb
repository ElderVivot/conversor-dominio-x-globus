```python
import pandas as pd
from typing import List, Dict
```

```python
dict_df = pd.read_excel(
    io="../data/_relacao_colaboradores_grupo_mobi.xls",
    sheet_name=None,
    usecols='A,D'
)
```

```python
dict_df
```

```python
df: pd.DataFrame = pd.concat([dict_df.get(nameSheet) for nameSheet in dict_df.keys()])
```

```python
df['CPF'] = df['CPF'].str.replace('.', '').str.replace('-', '')
```

```python
listCPFsPerCompanie: Dict[str, List[str]] = {}
for _, data in df.iterrows():
    codeCompanie = data['Empresa'][0:3]
    cpf = data['CPF']
    if cpf is None: continue
    try:
        listCPFsPerCompanie[codeCompanie].append(cpf)
    except Exception:
        listCPFsPerCompanie[codeCompanie] = [cpf]
```
